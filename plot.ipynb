```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Copy from W&B table
num_patches = [9, 100, 196, 324, 484, 625, 729, 841, 1024]
t1_acc = [0.98, 0.981, 0.976, 0.984, 0.98, 0.967, 0.948, 0.93, 0.88]
t5_acc = [0.997, 0.996, 0.986, 0.994, 0.991, 0.987, 0.96, 0.95, 0.93]
reg_acc = [0.976, 0.980, 0.979, 0.975, 0.978, 0.962, 0.941, 0.898, 0.827]

fig, ax = plt.subplots(figsize=(6, 4))
ax: plt.Axes
ax.plot(num_patches, t1_acc, marker='o', label='top-1')
ax.plot(num_patches, t5_acc, marker='x', label='top-5')
ax.plot(num_patches, reg_acc, color='red', marker='.', label='regression')


ax.set_ylim(0.751, 1.049)

ax.set_xlabel('Number of Patches')
ax.set_ylabel('Val Acc')
ax.set_title('Validation Accuracy vs Number of Patches')
ax.grid(True)
ax.legend()

fig.savefig('plots/val_acc_vs_num_patches.pdf', bbox_inches='tight', dpi=200)

fig.tight_layout()
plt.show()
```
